Did it always had the same popularity?

```python
import numpy as np

def reduce_ticks(num, n_ticks = 3):
    num /= 2
    div = 1
    while num > 9:
        num /= 10
        div *= 10
    return num * div

def popularity_fb_metrics(df):
    fig = plt.figure()
    likes_ax = fig.add_subplot(311)
    comments_ax = fig.add_subplot(312)
    shares_ax = fig.add_subplot(313)
    ts = pd.DataFrame(data = df[["fb_like","num_comments","shares"]])
    ts.index = pd.to_datetime(df["time_created"])
    ts.fb_like.plot(ax=likes_ax)
    ts.num_comments.plot(ax=comments_ax)
    ts.shares.plot(ax=shares_ax)
    likes_ax.set_ylabel('Likes')
    likes_ax.set_xlabel('Date')
    comments_ax.set_ylabel('Comments')
    comments_ax.set_xlabel('Date')
    shares_ax.set_ylabel('Shares')
    shares_ax.set_xlabel('Date')
    fig.set_tight_layout(True)
    
    maxval = max(ts.fb_like); ticks = np.arange(0, maxval, reduce_ticks(maxval))
    likes_ax.set_yticks(ticks)
    maxval = max(ts.num_comments); ticks = np.arange(0, maxval, reduce_ticks(maxval))
    comments_ax.set_yticks(ticks)
    maxval = max(ts.shares); ticks = np.arange(0, maxval, reduce_ticks(maxval))
    shares_ax.set_yticks(ticks)
    plt.show()
    
popularity_fb_metrics(dataset)
```

**Slicing dataframe for articles to be within from a month ago to the back**

```python
from datetime import datetime
import dateutil.relativedelta
a_month = dateutil.relativedelta.relativedelta(months=1)
end_time = str(pd.Timestamp("2016-08-14") - a_month)
end_time
```

```python
def slice_dataset(df, start_time, end_time):
    return df[(df['time_created'] >= start_time) & (df['time_created'] <= end_time)]
    
dataset_sliced = slice_dataset(dataset, "2015-01-01", end_time)
```

**Setting the most voted reaction as a label.**

```python
reactions_list = ["acamazing", "fail", "wtf", "lol", "trashy",
                  "win", "ew", "omg", "cute", "yaaass", "love",
                  "hate", "drab", "blimey", "fab", "ohdear",
                  "splendid"]

def highest_reaction(dataframe_row):
    max_votes = -1
    max_reaction = "NA"
    for reaction in reactions_list:
        reaction_votes = dataframe_row[reaction]
        if reaction_votes > max_votes:
            max_votes = reaction_votes
            max_reaction = reaction
    if max_reaction == "NA":
        print("NA found")
    return max_reaction

function = highest_reaction
dataset_sliced.loc[:,function.__name__] = dataset_sliced.apply(lambda x: function(x), axis=1)
```

```python
dataset_sliced[["name", "highest_reaction"]].head(2)
```

**Box plot**

```python
plt.rc('figure', figsize=(15,8))

def boxplots_fblike_reactions(df):    
    medians = (dataset_sliced[["fb_like", "highest_reaction"]]. \
                   groupby(["highest_reaction"]).median()). \
                        sort_values(["fb_like"], ascending=[False])
    ax = sns.boxplot(x="fb_like", y="highest_reaction", data=df, 
                     orient="h", order=medians.index)
    plt.xlim([-200,170000])
    plt.title("")
    plt.suptitle("")
    plt.ylabel("Reaction", fontsize=16)
    plt.xlabel("Likes", fontsize=16)
    plt.yticks(fontsize=14)
    plt.xticks(fontsize=14)
    plt.tight_layout()
    plt.show()

boxplots_fblike_reactions(dataset_sliced)
```

## Finding correlations between likes and each reaction

The first interesting question I wanted to answer was is there a linear correlation between the likes and the reaction?, if there is not, then there may be some of them more correlated than the others and those should be the most important.

First by using linear regressions.

```python
plt.rc('figure', figsize=(8,4))

def regplot(X, Y, x_label, y_label, angle=0):
    plt.scatter(X, Y, color='#0080ff', edgecolor='black', linewidth='0.5')
    sns.regplot(X, Y, scatter=False, color="#df2020")
    plt.ylim([0,5000])
    plt.title("%s vs %s" % (x_label, y_label), fontsize=18, y=1.05)
    plt.xlabel(x_label, fontsize=16)
    plt.ylabel(y_label, fontsize=16)
    plt.xticks(rotation=angle)
    plt.tight_layout()

plt.figure(1)
plt.subplot(131)
regplot(dataset_sliced["fb_like"], dataset_sliced["yaaass"], "yaaass", "likes", 70)
plt.subplot(132)
regplot(dataset_sliced["fb_like"], dataset_sliced["cute"], "cute", "likes", 70)
plt.subplot(133)
regplot(dataset_sliced["fb_like"], dataset_sliced["love"], "love", "likes", 70)
```

Apparently a love reaction could generate more likes, we could do even more preprocessing to see if we can get rid of unusual values or outliers.

**Adding an "is it a listcicle?" column**

```python
import nltk
def listcicle_number(dataframe_row):
    tokenizer = nltk.RegexpTokenizer(r"[a-zA-Z0-9']+")
    tokens = tokenizer.tokenize(dataframe_row["name"])
    for i in range(0, int(len(tokens)/2)):
        if tokens[i].isdigit():
            number = int(tokens[i])
            if '.' not in tokens[i]:
                return number
    return "NA"

function = listcicle_number
dataset_more_sliced.loc[:,function.__name__] = dataset_more_sliced.apply(lambda x: function(x), axis=1)
```

```python
dataset_more_sliced[['name', 'listcicle_number']].head(2)
```

```python
def only_listcicles(df):
    return df[(df['listcicle_number'] != 'NA')]

# dataset_more_sliced
dataset_msliced_listcicles = only_listcicles(dataset_more_sliced)
```

## Building the classifier

We want to predict the reaction a headline could produce, and we have limited this study on only textual features, there are certainly a bunch of other factors that could make more impact on the prediction like the thumbnail and the time, we didn't take them into account.

![Image](images/listcycle.png)

**Only headlines as features**

We figured out the dataset contains english and spanish headlines, spanish ones must be taken away from the corpus. But first we added a tokenized text column.

```python
from nltk.tokenize import RegexpTokenizer
def tokenized_name(dataframe_row):
    # To lowercase
    text = dataframe_row["name"].lower()
    # Tokenize words without punctuation
    tokenizer = nltk.RegexpTokenizer(r"[a-zA-Z']+")
    tokens = tokenizer.tokenize(text)
    return " ".join(tokens)

function = tokenized_name
dataset_msliced_listcicles.loc[:,function.__name__] = dataset_msliced_listcicles.apply(lambda x: function(x), axis=1)
```

## More visualization

**Histogram showing the number of votes for each reaction in descendent order.**

```python
plt.rc('figure', figsize=(10,7))

def histogram_reactions(df):

    counter_per_reaction = []
    for reaction in reactions_list:
        posts_with_reaction = df[df["highest_reaction"] == reaction]
        counter_per_reaction.append(len(posts_with_reaction))
    values, keys = zip(*sorted(zip(counter_per_reaction, reactions_list), reverse=True))
    ax = sns.countplot(y="highest_reaction", data=df, order=keys, linewidth = 3, color = '#0080ff')
    ax.set_xlim([0, 5600])
    for p, v in zip(ax.patches, values):
        ax.annotate(v, (p.get_width()+30, p.get_y()+0.6), size=14)
    plt.xlabel("Number of headlines", fontsize=16)
    plt.ylabel("Reaction", fontsize=16)
    plt.yticks(fontsize=14)
    plt.xticks(fontsize=14)
    plt.tight_layout()

    plt.show()
    
histogram_reactions(dataset_msliced_listcicles)
```

And there are a lot more love and lol reactions, so we get rid of the ones that have 0 samples and make new groups with the others to relate them with Facebook reactions.

According to the correspondence below made by looking at the last correlation matrix and by taking into account only high arousal emotions, so we are not considering "SAD" as a useful reaction.

`
love, splendid, fab - LOVE (0)
lol, fail -> HAHA (1)
win, cute, yaaass, omg, wtf, ohdear, acamazing, blimey -> WOW (2)
hate, trashy, ew, drab -> ANGRY (3)
`

## Features

Discretizing the class 

```python
fb_reactions_list = ["LOVE", "HAHA", "WOW", "ANGRY"]
```

```python
reaction_class = {"acamazing": 2, "fail": 1, "wtf": 3, "lol": 1, "trashy": 3,
                  "win": 2, "ew": 3, "omg": 2, "cute": 2, "yaaass": 2, "love": 0,
                  "hate": 3, "drab": 3, "blimey": 2, "fab": 0, "ohdear": 2,
                  "splendid": 0}

def highest_reaction_class(dataframe_row):
    return reaction_class[dataframe_row["highest_reaction"]]

def highest_reaction_classname(dataframe_row):
    return fb_reactions_list[reaction_class[dataframe_row["highest_reaction"]]]

function = highest_reaction_class
dataset_msliced_listcicles.loc[:,function.__name__] = dataset_msliced_listcicles.apply(lambda x: function(x), axis=1)
function = highest_reaction_classname
dataset_msliced_listcicles.loc[:,function.__name__] = dataset_msliced_listcicles.apply(lambda x: function(x), axis=1)
```

So now it looks like this

```python
plt.rc('figure', figsize=(10,2))

def histogram_fbreactions(df):

    counter_per_reaction = []
    for reaction in fb_reactions_list:
        posts_with_reaction = df[df["highest_reaction_classname"] == reaction]
        counter_per_reaction.append(len(posts_with_reaction))
    values, keys = zip(*sorted(zip(counter_per_reaction, fb_reactions_list), reverse=True))
    ax = sns.countplot(y="highest_reaction_classname", data=df, order=keys, linewidth = 3, color = '#0080ff')
    ax.set_xlim([0, 5700])
    for p, v in zip(ax.patches, values):
        ax.annotate(v, (p.get_width()+30, p.get_y()+0.55), size=14)
    plt.xlabel("Number of headlines", fontsize=16)
    plt.ylabel("Reaction", fontsize=16)
    plt.yticks(fontsize=14)
    plt.xticks(fontsize=14)
    plt.tight_layout()

    plt.show()
    
histogram_fbreactions(dataset_msliced_listcicles)
print("%s rows" % len(dataset_msliced_listcicles))
```
